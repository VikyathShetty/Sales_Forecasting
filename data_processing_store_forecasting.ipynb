```python
# Load datasets
train = pd.read_csv('train.csv') 
train
```

```python
# Load datasets
train = pd.read_csv('train.csv',parse_dates = ['date']) 
train
```

```python
test = pd.read_csv('test.csv')
test
```

```python
stores = pd.read_csv('stores.csv')
stores
```

```python
oil = pd.read_csv('oil.csv', parse_dates=['date'])
oil
```

```python
holidays_events = pd.read_csv('holidays_events.csv', parse_dates=['date'])
holidays_events
```

```python
print(train.isnull().sum())
print(30*"-")
print(test.isnull().sum())
print(30*"-")
print(stores.isnull().sum())
print(30*"-")
print(oil.isnull().sum())
print(30*"-")
print(holidays_events.isnull().sum())
```

```python
# Handle missing values in oil prices using interpolation
oil['dcoilwtico'] = oil['dcoilwtico'].interpolate(method='linear')
```

```python
# Merge datasets
train = train.merge(stores, on='store_nbr', how='left')
train = train.merge(oil, on='date', how='left')
train = train.merge(holidays_events, on='date', how='left')
```

```python
# Convert date columns to datetime format
train['date'] = pd.to_datetime(train['date'])
oil['date'] = pd.to_datetime(oil['date'])
holidays_events['date'] = pd.to_datetime(holidays_events['date'])
```

```python
# Extracting day week month and year
train['day'] = train['date'].dt.day
train['week'] = train['date'].dt.isocalendar().week
train['month'] = train['date'].dt.month
train['year'] = train['date'].dt.year
train['day_of_week'] = train['date'].dt.dayofweek
```

```python
train
```

```python
# Aggregate sales by month to identify seasonal patterns
monthly_sales = train.groupby(['year', 'month'])['sales'].sum().reset_index()

# Pivot data for better visualization
monthly_sales_pivot = monthly_sales.pivot(index='month', columns='year', values='sales')

# Plot the sales trend for each year
plt.figure(figsize=(12, 6))
sns.heatmap(monthly_sales_pivot, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Monthly Sales Trends Across Years")
plt.ylabel("Month")
plt.xlabel("Year")
plt.show()

#Darker area indicates higher sales. Sales are increasing year by year and also sales are more on the year end.
```

```python
# Holiday & Event Features
train['is_holiday'] = train['type_y'].notna().astype(int)
train['is_weekend'] = (train['day_of_week'] >= 5).astype(int)
train['is_gov_payday'] = train.apply(
    lambda row: int(row['day'] in [15, row['date'].days_in_month]), axis=1
)

train['earthquake_impact'] = (train['date'] == '2016-04-16').astype(int)
```

```python
train
```

```python
# Store-Specific Aggregations
store_avg_sales = train.groupby('store_nbr')['sales'].mean().rename('avg_store_sales')
train = train.merge(store_avg_sales, on='store_nbr', how='left')
```

```python
store_avg_sales
```

```python
# Rolling Statistics & Lag Features
train['sales_lag_7'] = train.groupby(['store_nbr', 'family'])['sales'].shift(7)
train['sales_lag_30'] = train.groupby(['store_nbr', 'family'])['sales'].shift(30)
train['rolling_mean_7'] = train.groupby(['store_nbr', 'family'])['sales'].rolling(7).mean().reset_index(level=[0,1], drop=True)
train['rolling_std_7'] = train.groupby(['store_nbr', 'family'])['sales'].rolling(7).std().reset_index(level=[0,1], drop=True)

```

```python
# This code identifies the top 5 best-selling product families within each store cluster.


top_products = train.groupby(['cluster', 'family'])['sales'].sum().reset_index()

top_products['rank'] = top_products.groupby('cluster')['sales'].rank(ascending=False, method='dense')

top_products = top_products[top_products['rank'] <= 5]
```

```python
top_products
```

```python
# Exploratory Data Analysis (EDA)
print("Exploratory Data Analysis")
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='sales', data=train, label='Sales Trend')
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.show()

```

```python
# Analyze Pre/Post-Holiday and Promotion Sales
holiday_sales = train.groupby(['date', 'is_holiday'])['sales'].mean().reset_index()
sns.boxplot(x='is_holiday', y='sales', data=holiday_sales)
plt.title('Sales Before and After Holidays')
plt.xlabel('Is Holiday')
plt.ylabel('Sales')
plt.show()
```

```python
# Check Correlations Between Oil Prices & Sales
correlation = train[['sales', 'dcoilwtico']].corr()
print("Correlation between Sales and Oil Prices:")
print(correlation)

```

```python
# EDA of oil price and sales
daily_sales = train.groupby('date')['sales'].sum().reset_index()

daily_sales = daily_sales.merge(oil, on='date', how='left')

fig, ax1 = plt.subplots(figsize=(12,6))

ax1.set_xlabel('Date')
ax1.set_ylabel('Total Sales', color='blue')
ax1.plot(daily_sales['date'], daily_sales['sales'], color='blue', label='Sales')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Oil Price (USD)', color='red')
ax2.plot(daily_sales['date'], daily_sales['dcoilwtico'], color='red', linestyle='dashed', label='Oil Price')
ax2.tick_params(axis='y', labelcolor='red')

fig.legend(loc="upper left")
plt.title("Oil Prices vs. Sales Over Time")
plt.show()
```

```python
# Detect Anomalies in Sales Trends
sns.boxplot(y='sales', data=train)
plt.title('Sales Anomalies Detection')
plt.ylabel('Sales')
plt.show()
```
